```python
# As we know pie chart can only be built on numerical or float values , so we now have total value  and now we will save it in a list and provide labels.
lst1 = [3407,7832]
label1 = ['M','F']
plt.pie(lst1 , labels= label1 , labeldistance = 0.8 , autopct = "%0.2f%%")
```

```python
# Now we want to know that , based on Gender who have spend more amount
Sales_gender = df.groupby(['Gender'] , as_index=False)['Amount'].sum().sort_values(by='Amount' , ascending = False)
Sales_gender
```

```python
x = Sales_gender['Amount']
y = Sales_gender['Gender']
plt.pie(x , labels = x , labeldistance = 1.3 , autopct="%0.2f%%")
plt.text(0.0,0.5,'Female' , color='white' , bbox={'facecolor':'y'})
plt.text(0.3,-0.2,'Male' , color='white' , bbox={'facecolor':'y'})
```

**Or we can also create a bar chart for this like**

```python
x1 = Sales_gender['Gender']
y1 = Sales_gender['Amount']
c= ['r','y']
plt.bar(x1,y1,color=c)

def showlabels(x,y):
    for i in range(len(x)):
        plt.text(i,y[i],y[i])

showlabels(x1,y1)

plt.show()
```

**From the above graph we can see that most of the buyers are females and even purchasing power of females are greater than men.**

**State**

```python
# Now , we want to find total number of orders from top 10 States.
df.columns
```

```python
sales_states = df.groupby(['State'] , as_index=False)['Orders'].sum().sort_values(by='Orders' , ascending = False).head(10)
sales_states
```

```python
x4 = sales_states['State']
y4 = sales_states['Orders']
c = ['r','y','g','m','r','b','r','y','g']
# Now all the states names are arranged in a very bad manner , so we have to increase plot size.
plt.figure(figsize=(20,5))
plt.bar(x4,y4,width=0.8,color=c)
plt.title('Orders by States',fontsize = 15)
plt.xlabel('Orders',fontsize=15)
plt.ylabel('State',fontsize=15)
```

```python
# Now if we want to find total amount spend by particular state.
amount_states = df.groupby(['State'] , as_index=False)['Amount'].sum().sort_values(by='Amount' , ascending = False).head(10)
amount_states
```

```python
x5 = amount_states['State']
y5 = amount_states['Amount']
c = ['r','y','g','m','r','b','r','y','g']
# Now all the states names are arranged in a very bad manner , so we have to increase plot size.
plt.figure(figsize=(20,5))
plt.bar(x5,y5,width=0.8,color=c)
plt.title('Amount spend by States',fontsize = 15)
plt.xlabel('Amount',fontsize=15)
plt.ylabel('State',fontsize=15)
```

**From the above two graphs we can configure about , total Orders by States and total amount spend by Different States**
